```python
#converto la variabile categoria che servirà per la y
data_df.loc[data_df["Churn"]=="Yes", "Churn"] = 1
data_df.loc[data_df["Churn"]=="No", "Churn"] = 0
```

```python
data_df.info()
```

```python
#individuo le colonne con valori mancanti e le cancello
data_df.replace('No internet service', np.nan , inplace=True)
data_df.replace('No phone service', np.nan , inplace=True)
data_df.replace(' ', np.nan , inplace=True)
```

```python
data_df.isnull().sum()
```

```python

data_df.drop(["MultipleLines", "OnlineBackup", "TechSupport", "OnlineSecurity",
           "DeviceProtection", "StreamingTV", "StreamingMovies"], axis=1, inplace=True)
```

```python
#separo target e predizioni e divido tra training e validation
from sklearn.model_selection import train_test_split

y = data_df.Churn

#CustomerID, paperlessBilling, paymentMethod non servono ai fini della predizione

features = ['gender','SeniorCitizen', 'Partner',
        'Dependents', 'tenure', 'PhoneService',
       'InternetService', 'Contract',
        'MonthlyCharges', 'TotalCharges']
X = data_df[features]

X_train_full, X_valid_full, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                                random_state=0)


low_cardinality_cols = [cname for cname in X_train_full.columns if X_train_full[cname].nunique() < 10 and 
                        X_train_full[cname].dtype == "object"]

# Seleziono le colonne con valori numerici
numerical_cols = [cname for cname in X_train_full.columns if X_train_full[cname].dtype in ['int64', 'float64']]


my_cols = low_cardinality_cols + numerical_cols
X_train = X_train_full[my_cols].copy()
X_valid = X_valid_full[my_cols].copy()

print(X_train.shape, y_train.shape, X_valid.shape, y_valid.shape)
```

```python
# lista delle variabili categoriche
s = (X_train.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
#variabili categoriche trasformate con Label Encoder: si assegna un integer diverso ad ogni valore
from sklearn.preprocessing import LabelEncoder

label_X_train = X_train.copy()
label_X_valid = X_valid.copy()

label_encoder = LabelEncoder()
for col in object_cols:
    label_X_train[col] = label_encoder.fit_transform(X_train[col])
    label_X_valid[col] = label_encoder.transform(X_valid[col])
```

```python
label_X_train.info()
```

```python
#utilizzo questo classificatore perchè permette di raggiungere un'accuratezza maggiore
#rispetto ad utilizzare un singolo albero di decisione, infatti
#vengono utilizzati più alberi e in ogni componente vengonon fatte delle predizioni
#infine si ha una votazione in cui viene presa la decisione considerata la migliore

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(label_X_train, y_train)
preds = forest_model.predict(label_X_valid)
print(mean_absolute_error(y_valid, preds))
```

```python
#accuratezza
from sklearn.metrics import accuracy_score

random_forest_acc_score = accuracy_score(preds.round(), y_valid) * 100
print("Accuracy Score:", random_forest_acc_score)
```
